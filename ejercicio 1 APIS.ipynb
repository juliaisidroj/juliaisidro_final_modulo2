```python
#Primer paso siempre
import requests #Permite coger los datos de la libreria
import pandas as pd
```

```python
def llamar_api (endpoint):
    datos = requests.get (endpoint)
    print (f"Llamada al API, con respuesta: {datos.status_code}")

    if datos.status_code != 200:
        print(f"Llamada fallo. por el siguiente motivo: {datos.reason}")
    else:
        return datos.json()
```
